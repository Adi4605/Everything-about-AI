```python
import numpy as np
```

### Binary Step Function

```python
def binary_step(x):
  return np.where(x>0,1,0)
```

### Linear Function

```python
def linear(x):
  return x
```

### Sigmoid Function

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

### Tanh Function

```python
def tanh1(x):
  return np.tanh(x)
```

```python
def tanh2(x):
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

### ReLU (Rectified Linear Unit)

```python
def relu(x):
  return np.maximum(0,x)
```

### Leaky ReLU

```python
def leaky_relu(x,alpha=0.01):
  return np.where(x>0, x, alpha*x)
```

### Parametric ReLU (PReLU)

```python
def prelu(x,alpha):
  return np.where(x>0, x, alpha*x)
```

### ELU (Exponential Linear Unit)

```python
def elu(x, alpha=1.0):
  return np.where(x>0, x, alpha*(np.exp(x)-1))
```

### SELU (Scaled ELU)

```python
def selu(x):
  alpha=1.6733
  scale=1.0507
  return scale*np.where(x>0, x, alpha*(np.exp(x)-1))
```

### Softplus

```python
def softplus(x):
  return np.log(x/(1+np.abs(x)))
```

### Softsign

```python
def softsign(x):
  return x/(1+np.abs(x))
```

### Swish Function

```python
def swish(x):
  return x*sigmoid(x)
```

### Mish Function

```python
def mish(x):
  return x*np.tanh(np.log(1+np.exp(x)))
```

### GELU (Gaussian Error Linear Unit)

```python
def gelu(x):
  return 0.5*x*(1+np.tanh(np.sqrt(2/np.pi)*(x+0.044715*np.power(x,3))))
```

### Softmax

```python
def softmax(x):
  e_x = np.exp(x-np.max(x)) #for numerical stability
  return e_x/e_x.sum(axis=0)
```
